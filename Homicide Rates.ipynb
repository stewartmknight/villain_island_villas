```python
# Dependencies
import pandas as pd
```

```python
# Read CSV
path = "Resources/homicide_data.csv"
homicide_df = pd.read_csv(path)
homicide_df.head()
```

```python
# Eliminate excess columns 
reduced_homicide_df = homicide_df[["Unnamed: 0", "Estimates of rates of homicides per 100 000 population"]]
reduced_homicide_df.head()
```

```python
# Rename columns
renamed_homicide_df = reduced_homicide_df.rename(columns = {"Unnamed: 0": "Country",
                                                          "Estimates of rates of homicides per 100 000 population": "Homicide Rate"})
renamed_homicide_df.head()
```

```python
# Clean old identifying information
dropped_homicide_df = renamed_homicide_df.drop([0, 1], axis = 0)
dropped_homicide_df.head()
```

```python
# Export new CSV
dropped_homicide_df.to_csv("Output/Homicide_Rate.csv", index=False, header=True)
```
